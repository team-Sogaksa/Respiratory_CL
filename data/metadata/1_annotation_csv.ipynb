```python
concat_df = []
all_combinations = {(0, 0), (0, 1), (1, 0), (1, 1)}  # 모든 조합을 세트로 정의

for idx, txt_file in enumerate(sorted(txt_files)):
    filename = txt_file.replace('.txt', '')
    df = pd.read_csv(os.path.join(database_dir, txt_file), sep='\t', header=None)
    df.columns = ['start', 'end', 'crackles', 'wheezes']
    df['filename'] = filename

    # 각 파일에서 나타나는 조합을 추적
    observed_combinations = set((row['crackles'], row['wheezes']) for index, row in df.iterrows())

    # 모든 조합이 있는지 확인
    if observed_combinations >= all_combinations:
        print(f"File {filename} contains all crackles and wheezes combinations.")
        break
    concat_df.append(df)
    # if idx == 2: break  # 필요에 따라 이 부분은 조정 가능

# combined_df = pd.concat(concat_df, ignore_index=True)
```

```python
file_paths = [os.path.join(database_dir, f) for f in sorted(txt_files)]
```

# save csv

txt 파일들 읽어서 모든데이터 리스트(all_data)에 담고

담은 리스트로 데이터프레임 생성, csv로 저장

```python
all_data = []  # 모든 파일의 데이터를 저장할 리스트

for file_path in file_paths:
    with open(file_path, 'r', encoding='utf-8') as file:
        data = [line.strip().split('\t') for line in file]
        filename = os.path.basename(file_path).replace('.txt', '')
        for entry in data:
            # 각 줄을 리스트로 처리하고 파일 이름을 추가
            all_data.append(entry + [filename])
```

```python
len(all_data)
```

```python
# DataFrame 생성
df = pd.DataFrame(all_data, columns=['start', 'end', 'crackles', 'wheezes', 'filename'])
df[['crackles', 'wheezes']] = df[['crackles', 'wheezes']].astype(int)

# 조건에 따라 추가 컬럼 계산
df['both'] = ((df['crackles'] == 1) & (df['wheezes'] == 1)).astype(int)
df['normal'] = ((df['crackles'] == 0) & (df['wheezes'] == 0)).astype(int)

# 최종 DataFrame 확인
df.head()
```

```python
cols = ['filename', 'start', 'end', 'crackles', 'wheezes', 'both', 'normal']
df = df[cols]
```
