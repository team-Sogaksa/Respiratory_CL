### get csv

```python
def counting_class(database_dir, file_path):
    # 파일 읽기
    df = pd.read_csv(file_path, header=None, delim_whitespace=True, names=['start', 'end', 'crackles', 'wheezes'])
    y, sr = librosa.load(os.path.join(database_dir, file_path.replace('.txt', '.wav')), sr=8000)
    duration = len(y)/sr

    # 카운트 딕셔너리
    counts = {
        'filename': os.path.basename(file_path).replace('.txt', ''),
        'duration': duration,
        'crackles': df['crackles'].sum(),
        'wheezes': df['wheezes'].sum(),
        'only_crackles': ((df['crackles'] == 1) & (df['wheezes'] == 0)).sum(),
        'only_wheezes': ((df['crackles'] == 0) & (df['wheezes'] == 1)).sum(),
        'both': ((df['crackles'] == 1) & (df['wheezes'] == 1)).sum(),
        'normal': ((df['crackles'] == 0) & (df['wheezes'] == 0)).sum()
    }
    return counts
```

```python
# 모든 파일의 결과를 리스트에 저장
# results = [counting_class(database_dir, os.path.join(database_dir, f)) for f in txt_files]
results = [counting_class(database_dir, os.path.join(database_dir, f)) for f in tqdm(txt_files, desc="Processing files")]

# 결과 리스트를 데이터프레임으로 변환
results_df = pd.DataFrame(results)

# 데이터프레임 출력
results_df
```
